```python
import pandas as pd
import requests
import certifi

url = "https://pokemondb.net/pokedex/all"

html = requests.get(url, verify=certifi.where()).text
df = pd.read_html(html, attrs={"id": "pokedex"})[0]

types = (
    df["Type"].astype(str)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
    .str.split(" ", n=1, expand=True)
)
df["Type 1"] = types[0]
df["Type 2"] = types[1]
df = df.drop(columns=["Type"])

df["#"] = df["#"].astype(str).str.zfill(4)

df.to_csv("pokemon_pokedex_all.csv", index=False, encoding="utf-8")
print("Saved: pokemon_pokedex_all.csv")
```
